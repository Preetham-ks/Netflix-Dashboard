```python
import pandas as pd
```

```python
df = pd.read_csv('netflix_titles.csv')
df.info()
df.head()
```

## check null values

```python
df.isnull().sum()
```

```python
df.dropna(subset=['title' , 'type' , 'release_year'], inplace=True)
```

```python
df['rating'].fillna('Not Rated', inplace= True)
df['country'].fillna('Unknown', inplace= True)
df['director'].fillna('Unknown', inplace= True)
df['cast'].fillna('Unknown', inplace= True)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
df['year_added'] = df['date_added'].dt.year
```

```python
df['date_added'] = df['date_added'].str.strip()
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
```

```python
df['date_added'] = df['date_added'].astype(str).str.strip()

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

df['year_added'] = df['date_added'].dt.year
```

```python
df['genres'] = df['listed_in'].str.split(', ')
```

```python
df['main_country'] = df['country'].apply(lambda x: x.split(',')[0])
```

```python
df.info()
df.head()
```

```python
df.isnull().sum()  
df['duration'].unique() 
```

```python
df['duration_minutes'] = df['duration'].str.extract(r'(\d+)\s*min').astype(float)
df['duration_seasons'] = df['duration'].str.extract(r'(\d+)\s*Season').astype(float)
```

## EDA Process (Exploratory Data Analysis)

```python
df.describe(include='all') 
df['type'].value_counts()  
df['rating'].value_counts() 
```

```python
df['title'].value_counts().head(10)  
df['main_country'].value_counts().head(10) 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(data=df, x='type', hue='type', palette='pastel',legend=False)
plt.title('Movies vs TV Shows on Netflix')
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot(kind='line', figsize=(10, 5))
plt.title('Total Titles by Release Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
top_countries = df['main_country'].value_counts().head(5)

top_countries.plot(kind='barh', color='teal')
plt.title('Top Countries by Number of Netflix Titles')
plt.xlabel('Number of Titles')
plt.show()
```

```python
top_directors = df['director'].value_counts().head(10)

top_directors.plot(kind='barh', color='purple')
plt.title('Top 10 Directors on Netflix')
plt.xlabel('Number of Titles')
plt.show()
```

```python
# Remove missing or invalid director names
df_valid_directors = df[df['director'].notna()]
df_valid_directors = df_valid_directors[df_valid_directors['director'] != 'Not Rated']
```

```python
top_directors = df_valid_directors['director'].value_counts().head(10)
top_directors.plot(kind='barh', color='purple')
plt.title('Top 10 Directors on Netflix')
plt.xlabel('Number of Titles')
plt.show()
```

```python
from collections import Counter

genre_list = sum(df['genres'].dropna(), [])  # flatten list
top_genres = Counter(genre_list).most_common(10)

genres_df = pd.DataFrame(top_genres, columns=['Genre', 'Count'])
sns.barplot(x='Count', y='Genre', data=genres_df, palette='coolwarm')  
plt.title('Top 10 Genres on Netflix')
plt.show()
```
